```python
import pandas as pd

df = pd.read_csv("/kaggle/input/ibm-hr-analytics-attrition-dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

```python
df.shape
df.info()
df.describe(include="all")
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)
df.head(2)
```

```python
cat_cols = ["attrition","department","educationfield","gender","jobrole","maritalstatus","overtime"]
for c in cat_cols:
    df[c] = df[c].astype("category")

df.dtypes.head(10)
```

```python
df["attrition_flag"] = (df["attrition"] == "Yes").astype(int)
df["overtime_flag"] = (df["overtime"] == "Yes").astype(int)

# Altersgruppen
bins_age   = [17,25,35,45,55,80]
labels_age = ["18-25","26-35","36-45","46-55","56+"]
df["age_band"] = pd.cut(df["age"], bins=bins_age, labels=labels_age)

# Firmenzugehörigkeit (Tenure)
bins_ten   = [-1,1,3,5,10,40]
labels_ten = ["≤1","2-3","4-5","6-10","10+"]
df["tenure_band"] = pd.cut(df["yearsatcompany"], bins=bins_ten, labels=labels_ten)

# Einkommensband
bins_inc   = [-1,3000,6000,10000,20000]
labels_inc = ["≤3k","3-6k","6-10k","10k+"]
df["income_band"] = pd.cut(df["monthlyincome"], bins=bins_inc, labels=labels_inc)
```

```python
df[["attrition","attrition_flag","overtime","overtime_flag","age_band","tenure_band","income_band"]].head(10)
```
